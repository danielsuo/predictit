```python
import requests
import json
import re

import numpy as np
```

```python
r = requests.get("https://www.predictit.org/api/marketdata/all/")
```

```python
data = json.loads(r.content.decode("utf-8"))
```

```python
data["markets"]
```

```python
p = re.compile("Which party will win [\s\w\-]* in the 2020 presidential election")
```

```python
ids = []
markets = {}
for market in data["markets"]:
    if p.match(market["name"]):
        print(market["name"])
        ids.append(market["id"])
        
        markets[market["id"]] = market
        markets[market["id"]]["party"] = []
        markets[market["id"]]["price"] = []
        
        for contract in market["contracts"]:
            markets[market["id"]]["party"].append(contract["shortName"])
            markets[market["id"]]["price"].append(contract["lastTradePrice"])
            
        markets[market["id"]]["price"] = np.array(markets[market["id"]]["price"])
        markets[market["id"]]["price"] /= markets[market["id"]]["price"].sum()
```

```python
num_sims = 100
num_draws = 500

for id in markets.keys():
    ev = np.zeros((len(markets[id]["price"]), num_sims))
    for sim in range(num_sims):
        draws = np.random.choice(markets[id]["party"], num_draws, p=markets[id]["price"])
        for i, party in enumerate(markets[id]["party"]):
            result = (draws == party).sum()
            ev[i, sim] += result / num_draws
    
    print(markets[id]["shortName"], markets[id]["party"], ev.mean(axis=1), ev.std(axis=1))
```
